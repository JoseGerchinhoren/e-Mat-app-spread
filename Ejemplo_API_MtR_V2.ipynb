#Token de autenticación

```python
#Obtener el token de autenticación:

import requests
import json

#solicitud del access token a la api mtr
url = 'https://api.matbarofex.com.ar/v2/token/'
body = '{{"username":"{}","password":"{}"}}'.format(username, password)
headers = {'Content-Type': 'application/json'}

response = requests.post(url, data=body, headers=headers)

# Verificar el código de estado de la respuesta
if response.status_code == 200:
    # La solicitud fue exitosa
    data = response.json()
    # Procesar la respuesta JSON
    print(data)
else:
    # Hubo un error en la solicitud
    print('Error:', response.status_code)


token=data.get("access")
token='Bearer '+token
```

## Paso 4: Obtener Precios Tick by Tick

```python
from datetime import datetime, timedelta

# Obtener la fecha y hora actual en el formato requerido por la API
fecha_hoy = datetime.utcnow()
fecha_inicio = (fecha_hoy - timedelta(days=1)).strftime('%Y-%m-%dT%H:%M:%S.000Z')
fecha_fin = (fecha_hoy + timedelta(days=1)).strftime('%Y-%m-%dT%H:%M:%S.000Z')

# Definir los parámetros para la consulta de precios tick by tick
params_tick_prices = {
    'symbol': 'MAI.ROS/JUL24',  # Asegúrate de que este símbolo sea correcto
    'from': fecha_inicio,
    'to': fecha_fin,
    'pageSize': 100,
    'sortDir': 'desc',
    'market': 'matbarofex',
    'version': 'v2'
}

# Definir la URL del endpoint de precios tick by tick
url_tick_prices = 'https://apicem.matbarofex.com.ar/api/v2/tick-prices'

# Realizar la solicitud a la API para precios tick by tick
response_tick_prices = requests.get(url_tick_prices, headers=headers, params=params_tick_prices)

# Verificar el código de estado de la respuesta
print("Estado de la respuesta de precios tick by tick:", response_tick_prices.status_code)

# Intentar decodificar la respuesta como JSON
try:
    data_tick_prices = response_tick_prices.json()
    if 'content' in data_tick_prices and len(data_tick_prices['content']) > 0:
        tick_prices = data_tick_prices['content']
        print("Datos de precios tick by tick recientes:")
        print(json.dumps(tick_prices, indent=4))
        
        # Obtener los valores de compra, venta y último
        compra = tick_prices[0]['bidPrice'] if 'bidPrice' in tick_prices[0] else 'N/A'
        venta = tick_prices[0]['askPrice'] if 'askPrice' in tick_prices[0] else 'N/A'
        ultimo = tick_prices[0]['lastPrice'] if 'lastPrice' in tick_prices[0] else 'N/A'
        
        print(f"Instrumento: Maíz Ros Jul 24\nCompra: {compra}\nVenta: {venta}\nÚltimo: {ultimo}")
    else:
        print("No se encontraron precios tick by tick recientes.")
except json.JSONDecodeError:
    print("La respuesta de precios tick by tick no contiene un JSON válido:")
    print(response_tick_prices.text)
```

```python
#METODO WEBSOCKET

list_symbols = ['I.BTC', 'I.SOJA', 'I.AGTKSOYA']


import websocket
import json

# URL del WebSocket
url = 'wss://api.matbarofex.com.ar/v2/live/symbol/'

# Configuración de encabezado
headers = {
    'Authorization': token #'Bearer [access token]'
}

# Campos del mensaje de suscripción
subscription_message = {
    'action': 'subscribe',
    'request_id': 1,
    'symbols': list_symbols
}

# Función para manejar los eventos del WebSocket
def on_message(ws, message):

    print('Mensaje recibido:', message)

def on_error(ws, error):
    print('Error:', error)

def on_close(ws):
    print('Conexión cerrada')

def on_open(ws):
    print('Conexión abierta')
    # Envío del mensaje de suscripción al WebSocket
    ws.send(json.dumps(subscription_message))

# Crear una instancia de WebSocket y configurar los eventos
ws = websocket.WebSocketApp(url, header=headers)
ws.on_message = on_message
ws.on_error = on_error
ws.on_close = on_close
ws.on_open = on_open

# Iniciar la conexión WebSocket
ws.run_forever()
```
